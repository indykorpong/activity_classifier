```python
import numpy as np
import pandas as pd
import csv
import math
```

```python
def calc_sec(time):
    hms = time.split(':')
    hms = [float(x) for x in hms]
    sec = hms[2] + hms[1]*60 + hms[0]*3600
    sec = round(sec,3)
    return sec
```

```python
def calc_ts(sec):
    ts = ''
    hr = int(sec/3600)
    mn = int((sec - (hr*3600))/60)
    sc = sec - (hr*3600) - (mn*60)
    sc = round(sc,3)
    ts += str(hr) + ':' + str(mn) + ':' + str(sc)
    # print(ts)
    return ts
```

## Bai's Equation

```python
import matplotlib.pyplot as plt
```

```python
std_i_bar = [0.00349329,0.00465817,0.00543154]
std_i_bar = np.array(std_i_bar)

print(std_i_bar)
```

```python
def equation_bi(X_i):
    all_std = []
    
    std_i = np.std(X_i,axis=0)
    diff_std = std_i**2 - std_i_bar**2
    diff_std = (diff_std + 1) / (std_i_bar**2 + 1)
#     print(diff_std)

    all_std.append(diff_std)
    
    all_std = np.array(all_std)
#     print(all_std)
    
    ai = np.sum(all_std**2,axis=1)/3
    ai[ai<0] = 0
    ai = np.sqrt(ai)
#     print(ai)
    
    return ai
```

```python
def calc_t_period(new_lines,start_sec,prev_sec,prev_dat,last_sec):
    t_period = []
    for line in new_lines:
        dat = line[0]
        tim = line[1]

        sec = calc_sec(tim)
        diff_sec = sec-prev_sec

        if(diff_sec>3 and dat==prev_dat):
            t_period.append([dat,start_sec,prev_sec])
            start_sec = sec
        elif(dat!=prev_dat):
            t_period.append([prev_dat,start_sec,prev_sec])
            start_sec = sec
            prev_dat = dat
        elif(sec==last_sec):
            t_period.append([dat,start_sec,sec])

        prev_sec = sec
    return t_period
```
