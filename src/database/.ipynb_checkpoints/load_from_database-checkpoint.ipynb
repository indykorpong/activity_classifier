```python
import requests

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "http://122.155.202.69:1180/cu_amd/api/get_log.php"

data_length = 100
timestamp_from = '2019-02-01 23:41:06.000'
timestamp_to = '2019-02-28 23:41:06.000'

acc_querystring = {"log_type": "accelerometer_log",
               "offset": "0",
               "range": str(data_length),
               "user_id": "all",
               "filter_event_timestamp_from": timestamp_from,
               "filter_event_timestamp_to": timestamp_to}

payload = ""
headers = {
    'Authorization': "admin",
    'cache-control': "no-cache",
    'Postman-Token': "227e3a2e-4cbb-42bd-8c92-333088c70030"
    }

acc_response = requests.request("GET", url, data=payload, headers=headers, params=acc_querystring)
```

```python
json_acc = acc_response.json()

acc_length = json_acc['full_result_count']
acc_data = json_acc['data']
```

```python
hr_querystring = {"log_type": "hr_log",
                  "offset": "0",
                  "range": str(data_length),
                  "user_id": "all",
                  "filter_event_timestamp_from": timestamp_from,
                  "filter_event_timestamp_to": timestamp_to}

hr_response = requests.request("GET", url, data=payload, headers=headers, params=hr_querystring)
```

```python
json_hr = hr_response.json()

hr_length = json_hr['full_result_count']
hr_data = json_hr['data']
```
